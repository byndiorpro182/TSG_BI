```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.image as mpimg
```

```python
df = pd.read_excel('sale.xlsx','SalesOrders')
df['year'] = pd.DatetimeIndex(df['OrderDate']).year
df['month'] = pd.DatetimeIndex(df['OrderDate']).month
df['day'] = pd.DatetimeIndex(df['OrderDate']).day
df['length Of OrderDate'] = df['OrderDate'].astype(str).str.len()
df['OrderDateNew'] =  df['OrderDate'].astype(str).str[:4] + df['OrderDate'].astype(str).str[5:7] + df['OrderDate'].astype(str).str[8:10]
df.info()
df
#df['Region'].drop_duplicates()
```

```python
plt.figure(1,figsize=(25, 25))

plt.subplot(331)
plt.bar(df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Region'], df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Unit Cost'])
plt.ylabel('Unit Cost', fontsize = 15, color = 'blue')
plt.xlabel('Region', fontsize = 15, color = 'blue')
plt.title('This is title')
plt.text(0.65,300 ,'this is text',fontsize=30)
plt.grid(True)

plt.subplot(332)
plt.plot(df['OrderDate'][df['Region'] == 'West'],df['Unit Cost'][df['Region'] == 'West'])
plt.legend('West')
plt.plot(df['OrderDate'][df['Region'] == 'Central'],df['Unit Cost'][df['Region'] == 'Central'], 'y|--', lw = 5)
plt.legend('Central')
plt.plot(df['OrderDate'][df['Region'] == 'East'],df['Unit Cost'][df['Region'] == 'East'])
plt.legend('East')

plt.subplot(333)
plt.scatter(df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Region'], df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Unit Cost'])

plt.subplot(334)
plt.plot(df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Region'], df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Unit Cost'])

plt.subplot(335)
sizes = df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Unit Cost']
explode = (0, 0.1, 0)
labels = df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Region']
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle= -30)

plt.subplot(336)
plt.table(cellText=df[['Region','Rep','Unit Cost']].head(20).values, colLabels=df.columns, loc='center', fontsize = 30)
plt.subplot(337)
img = mpimg.imread('image 3840x2160.jpg')
imgplot = plt.imshow(img)
plt.subplot(338)
plt.barh(df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Region'], df[['Region','Unit Cost']].groupby('Region').sum().reset_index()['Unit Cost'])
plt.subplot(339)

plt.show()
```
